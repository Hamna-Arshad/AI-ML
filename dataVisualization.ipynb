```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = sb.load_dataset("iris")
#shape
print(f"Shape: {df.shape}")
```

```python
df.head(5)
```

```python
#information
df.info()
```

```python
#describe
df.describe()
```

```python
#visualization-scatter plots
sb.scatterplot(x='sepal_length', y='petal_length', data=df)
plt.title('Scatter Plot-Length')
plt.show()
```

```python
sb.scatterplot(x='sepal_width', y='petal_width', data=df)
plt.title('Scatter Plot-Width')
plt.show()
```

```python
#histogram
sb.histplot(data=df, x='petal_width', hue='species',multiple='stack')
plt.title('Histogram of Sepal Length')
plt.show()

sb.histplot(data=df, x='sepal_width', hue='species',multiple='stack')
plt.title('Histogram of Sepal Width')
plt.show()

sb.histplot(data=df, x='petal_length', hue='species',multiple='stack')
plt.title('Histogram of Petal Length')
plt.show()

sb.histplot(data=df, x='petal_width', hue='species',multiple='stack')
plt.title('Histogram of Petal Width')
plt.show()
```

```python
sb.boxplot(x='sepal_width', y='petal_width', data=df)
plt.title('Box Plot of Petal vs Sepal Width')
plt.show()

sb.boxplot(x='sepal_length', y='petal_length', data=df)
plt.title('Box Plot of Petal vs Sepal Length')
plt.show()

```
