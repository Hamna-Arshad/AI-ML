```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

endpoint = "https://models.github.ai/inference"
model = "meta/Llama-4-Scout-17B-16E-Instruct"
token = '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'

system_prompt = (
    "You are a kind and knowledgeable medical assistant. "
    "You answer general health-related questions in a clear and friendly way. "
    "You never diagnose conditions, prescribe medications, or give medical dosages. "
    "Always advise the user to consult a licensed medical professional for personal advice."
    "Donot give answers to anything other than medical questions and clearly mention that you are a medical assistant only."
)
client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)
```

```python
def is_safe(msg):
    unsafe_keywords = [
        "dosage", "overdose", "prescription", "suicide", "kill myself",
        "self-harm", "baby", "infant", "pregnant", "chest pain", "can't breathe",
        "heart attack", "stroke", "emergency", "antibiotic dose", "xanax",
        "prozac", "abortion", "illegal", "fake symptoms"
    ]
    msg = msg.lower()
    return not any(word in msg for word in unsafe_keywords)
```

```python
usrmsg = ''
while True:
    if usrmsg.lower() in['exit','quit']:
        print('Stay Healthy! GoodBye!!')
        break
    
    usrmsg = input("Question: ")
    if not is_safe(usrmsg):
        print(f'Question: {usrmsg}')
        print("I'm really sorry, but I'm not qualified to give advice on that. Please consult a doctor or a licensed medical professional.")
        print('\n\n')
        continue
    response = client.complete(
        messages=[
            SystemMessage(system_prompt),
            UserMessage(usrmsg),
        ],
        temperature=0.8,
        top_p=0.1,
        max_tokens=2048,
        model=model
    )
    print(f'Question: {usrmsg}')
    print(f'Response: {response.choices[0].message.content}')
    print('\n\n')
```
